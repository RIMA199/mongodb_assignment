```python
import pymongo
```

```python
import pandas as pd
```

```python
# Making a Connection with MongoClient
dbconn = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
# Database Name
dbname='ineuron'
```

```python
db=dbconn[dbname]
```

```python
#colleation name
collection_name = 'admission_details'
```

```python
# collection created
collection = db[collection_name]
```

```python
#read dataframe 
dataframe=pd.read_csv("https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv")
```

```python
dataframe.head()
```

```python
#create dictionary
data_dict = dataframe.to_dict("records")
```

```python
#insert records in a collection
collection.insert_many(data_dict)
```

```python
result_dataframe=collection.find()
```

```python
for i in result_dataframe:
    print(i)
```
